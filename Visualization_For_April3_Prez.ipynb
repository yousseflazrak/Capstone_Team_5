This is for visulization of the model results for April 3rd presentation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Load the model results for actor-critic
stf_ac = pd.read_csv("/Users/Simon/Desktop/shortfall_ac.csv")
```

```python
stf_ac["Shortfall"]
```

```python
sns.set()
ax = sns.lineplot(data = stf_ac)
ax.set_ylabel('Expected Shortfall',fontsize = 18)
ax.set_xlabel('Episodes ',fontsize = 18)
ax.set_xticklabels([0,0,2000,4000,6000,8000,10000],fontsize = 12)
ax.grid(b=True, which='major', color='white', linewidth=0.5,linestyle = '-')
```
